```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("billionares.csv")
```

```python
print(data.head())
```

```python
data.describe()
```

```python
# checking if any table has empty values
print(data.isnull().sum())
```

```python
# dropping the null values
data = data.dropna()
```

```python
# removing $,B and making all networths as floats
data["NetWorth"] = data["NetWorth"].str.strip("$")
data["NetWorth"] = data["NetWorth"].str.strip("B")
data["NetWorth"] = data["NetWorth"].astype(float)
```

```python
data
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Creating bar plots and scatter plots to understand the variation of net worth with the age of a person.

```python
plt.title("Distribution of Ages among the billionaires")
plt.hist(data.Age,bins = np.arange(20,100,10));
```

```python
plt.figure(figsize=(20, 10))
plt.title("Networth(in billions) with Age")
sns.scatterplot(x = 'Age',y = 'NetWorth',data = data,s=25);
```

#### Q1: Who are the ten most richest people in the world?

```python
df = data.sort_values(by = ["NetWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x="Name", hue="NetWorth", data=df)
plt.show()
```

The above plot shows the richest people in order, with Jeff Bezos at first, followed by Elon Musk and so on, with Mukesh Ambani at the tenth spot.

#### Q2: What are the top 5 domains with the most amount of billionares?

```python
a = data["Source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["blue", "orange", 'yellow', "green", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Domains to Become a Billionaire", fontsize=20)
plt.show()
```

This plot shows that these above domains are the best in order to become a billionaire.

#### Q3: What are the top 5 industries with the most amount of billionares?

```python
a = data["Industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["blue", "purple", 'yellow', "green", "pink"]
plt.figure(figsize=(5, 5))
plt.pie(industries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with Most Number of Billionaires", fontsize=20)
plt.show()
```

This plot shows that most of the billionaires on this list are from the above shown industries.

#### Q4: What are the top 5 countries with the most amount of billionares?

```python
a = data["Country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["blue", "purple", 'yellow', "green", "red"]
plt.figure(figsize=(5, 5))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```

This plot shows that most billionaires are citizens of these above mentioned countries.

#### Q5: Does age have any effect on the networth of an individual?

```python
plt.figure(figsize=(20,10))
plt.xlabel("Networth")
plt.ylabel("Age")
plt.plot(data.NetWorth,data.Age);
```

Here we can see that billionaires in this dataset are all from different ages. However, using this plot and the ones in the previous section we can reasonably assume that most billionaires are in the 50-80 age group.
